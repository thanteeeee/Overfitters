```python
import pandas as pd
import joblib
from google.colab import drive
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import FunctionTransformer

# 1. เชื่อมต่อ Drive
drive.mount('/content/drive')

# 2. ฟังก์ชัน Winsorize (ต้องนิยามไว้เหมือนเดิม)
def winsorize_data(X):
    X_copy = X.copy()
    if isinstance(X_copy, pd.DataFrame):
        for col in X_copy.columns:
            if len(X_copy[col]) > 0:
                X_copy[col] = winsorize(X_copy[col], (0.1, 0.1))
    return X_copy

# 3. โหลด Pipeline ที่บันทึกไว้
pipeline_path = '/content/drive/MyDrive/datasetProjectSleep/sleep_data_pipeline.joblib'
loaded_pipeline = joblib.load(pipeline_path)

# 4. โหลดข้อมูล (สมมติว่าเป็นข้อมูลใหม่)
# เราจะใช้ features เดิมที่เคยเตรียมไว้ (ไม่รวม target)
features_to_load = ['Age', 'Sleep Hours', 'Daily Screen Time', 'Stress Level', 'Use Before Sleep_enc', 'Anxiety/Low Mood_enc', 'Wellness Apps_enc', 'Feel Rested_enc', 'Sleep Quality_enc']
new_df = pd.read_csv('/content/drive/MyDrive/datasetProjectSleep/encoded_sleep_data_ready.csv')[features_to_load]

# 5. ใช้งาน Pipeline
X_new_transformed = loaded_pipeline.transform(new_df)

print("Preprocessing ด้วย loaded_pipeline สำเร็จ!")
# แสดงผลลัพธ์ 5 แถวแรก
display(pd.DataFrame(X_new_transformed, columns=features_to_load).head())
```
